```python
import pandas as pd
```

```python
# Leo Messi's Stats Breakdown
```

```python
messi_dataset=pd.read_excel("MessivsRonaldo_dataset.xls",sheet_name=("Messi"))

messi_dataset
```

```python
#Convert Columns to Numeric
```

```python
messi_dataset[["Appearances", "Minutes", "Goals", "Assists", "Key Passes Per Game ", "Successful Dribbles","Average Match Rating" ]] = messi_dataset[["Appearances", "Minutes", "Goals", "Assists", "Key Passes Per Game ", "Successful Dribbles","Average Match Rating"]].apply(pd.to_numeric)
```

```python
#Create Goals+Assists Column
```

```python
messi_dataset["Goals+Assists"] = messi_dataset["Goals"] + messi_dataset["Assists"]

messi_dataset
```

```python
# Goals per 90
```

```python
messi_dataset["Goals per 90"] = (messi_dataset["Goals"] / messi_dataset["Minutes"]) * 90

messi_dataset
```

```python
# Assists per 90
```

```python
messi_dataset["Assists per 90"] = (messi_dataset["Assists"] / messi_dataset["Minutes"]) * 90

messi_dataset
```

```python
# Distinguish between League and Champions League Competition
```

```python
messi_dataset['Competition'] = messi_dataset['Competition'].replace("SLL", "League")

messi_dataset
```

```python
# Change Column Order
```

```python
messi_dataset = messi_dataset.reindex(columns = ['Season','Team','Competition','Appearances','Minutes','Goals','Assists','Goals+Assists','Goals per 90','Assists per 90','Key Passes Per Game ','Shots Per Game','Successful Dribbles','Average Match Rating'])

```

```python
messi_dataset
```

```python
# Cristiano Ronaldo Stats Breakdown
```
